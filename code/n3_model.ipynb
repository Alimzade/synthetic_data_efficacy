```python
import os
import random
from PIL import Image

def random_sample_images(input_dir, output_dir, num_samples_per_class=10):
    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # List all class folders in the input directory
    class_folders = [folder for folder in os.listdir(input_dir) if os.path.isdir(os.path.join(input_dir, folder))]

    # Loop through each class folder
    for class_folder in class_folders:
        class_path = os.path.join(input_dir, class_folder)
        output_class_path = os.path.join(output_dir, class_folder)

        # Create the output subdirectory for the current class
        os.makedirs(output_class_path, exist_ok=True)

        # List all image files in the class folder
        image_files = [file for file in os.listdir(class_path) if file.lower().endswith(('.png', '.jpg', '.jpeg'))]

        # Randomly sample images from the current class
        sampled_images = random.sample(image_files, min(num_samples_per_class, len(image_files)))

        # Copy the sampled images to the output directory
        for image_file in sampled_images:
            input_image_path = os.path.join(class_path, image_file)
            output_image_path = os.path.join(output_class_path, image_file)

            # Open the image using Pillow and save it as PNG
            image = Image.open(input_image_path)
            image.save(output_image_path, format='PNG')

if __name__ == "__main__":
    # Set your input directory (contains folders for each class)
    input_directory = "c:\\Users\\user1\\Desktop\\data science lab\\topic03\\data\\Training"


    # Set your output directory (where you want to save the sampled PNG images)
    output_directory = "c:\\Users\\user1\\Desktop\\data science lab\\topic03\\data\\SampledTraining"

    # Set the number of samples you want per class
    num_samples_per_class = 10

    # Call the function to perform random sampling
    random_sample_images(input_directory, output_directory, num_samples_per_class)
```

```python
import os
import random
import SimpleITK as sitk

def bias_field_correction(input_image_path, output_image_path):
    # Read the input image
    input_image = sitk.ReadImage(input_image_path)

    # Check if the pixel type is float, if not, cast it
    if input_image.GetPixelIDType() != sitk.sitkFloat32:
        input_image = sitk.Cast(input_image, sitk.sitkFloat32)

    # Apply N4 correction
    n4_corrector = sitk.N4BiasFieldCorrectionImageFilter()
    output_image = n4_corrector.Execute(input_image)

    # Write the corrected image to the specified output path
    sitk.WriteImage(output_image, output_image_path)

def random_sample_and_correct_images(input_dir, output_dir, num_samples_per_class):
    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # List of classes
    classes = os.listdir(input_dir)

    for class_name in classes:
        class_path = os.path.join(input_dir, class_name)
        output_class_path = os.path.join(output_dir, class_name)

        # Skip non-directory entries
        if not os.path.isdir(class_path):
            continue

        # Create the output class directory if it doesn't exist
        os.makedirs(output_class_path, exist_ok=True)

        # List all image files in the class directory
        image_files = [f for f in os.listdir(class_path) if f.lower().endswith('.png')]

        # Randomly sample images
        sampled_images = random.sample(image_files, min(num_samples_per_class, len(image_files)))

        for image_file in sampled_images:
            # Construct the full paths for input and output images
            input_image_path = os.path.join(class_path, image_file)
            output_image_path = os.path.join(output_class_path, image_file)

            # Perform bias field correction
            bias_field_correction(input_image_path, output_image_path)

# Example usage:
input_directory = "c:\\Users\\user1\\Desktop\\data science lab\\topic03\\data\\Training"
output_directory = "c:\\Users\\user1\\Desktop\\data science lab\\topic03\\data\\biasTraining"
num_samples_per_class = 10

random_sample_and_correct_images(input_directory, output_directory, num_samples_per_class)
```
