```python
import matplotlib.pyplot as plt

# Data for Initial Classifier
initial_classifier = {
    'Accuracy': 92,
    'Precision': [0.94, 0.81, 0.97, 0.95],
    'Recall': [0.89, 0.83, 0.99, 0.96],
    'F1-Score': [0.91, 0.82, 0.98, 0.96]
}

# Data for Fine-Tuned Classifier
fine_tuned_classifier = {
    'Accuracy': 92,
    'Precision': [0.98, 0.77, 0.97, 0.95],
    'Recall': [0.83, 0.88, 0.99, 0.97],
    'F1-Score': [0.90, 0.82, 0.98, 0.96]
}

# Classes
classes = ['glioma', 'meningioma', 'notumor', 'pituitary']

# Plotting
plt.figure(figsize=(12, 8))

# Line Plot for Precision
plt.plot(classes, initial_classifier['Precision'], marker='o', linestyle='-', label='Precision before syntheetic data', color='yellow')
plt.plot(classes, fine_tuned_classifier['Precision'], marker='o', linestyle='--', label='Precision after syntheetic data' ,color='orange')

# Line Plot for Recall
plt.plot(classes, initial_classifier['Recall'], marker='o', linestyle='-', label='Recall before syntheetic data' , color='gray')
plt.plot(classes, fine_tuned_classifier['Recall'], marker='o', linestyle='--', label='Recall after syntheetic data', color="black")

# Line Plot for F1-Score
plt.plot(classes, initial_classifier['F1-Score'], marker='o', linestyle='-', label='F1-Score before synthetic data', color = 'red')
plt.plot(classes, fine_tuned_classifier['F1-Score'], marker='o', linestyle='--', label='F1-Score after synthetic data', color="brown")

plt.title('Comparison of Classifiers before synthetic data and after')
plt.xlabel('Classes')
plt.ylabel('Metric Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for Initial Classifier
initial_classifier = {
    'Accuracy': 93,
    'Precision': [0.93, 0.83, 0.98, 0.98],
    'Recall': [0.91, 0.88, 0.99, 0.94],
    'F1-Score': [0.93, 0.86, 0.98, 0.96]
}

# Data for Fine-Tuned Classifier
synthetic_classifier = {
    'Accuracy': 78,
    'Precision': [0.91, 0.53, 0.89, 0.87],
    'Recall': [0.67, 0.79, 0.91, 0.77],
    'F1-Score': [0.77, 0.63, 0.90, 0.82]
}

# Classes
classes = ['glioma', 'meningioma', 'notumor', 'pituitary']

# Plotting
plt.figure(figsize=(12, 8))

# Line Plot for Precision
plt.plot(classes, initial_classifier['Precision'], marker='o', linestyle='-', label='Initial Precision', color='yellow')
plt.plot(classes, synthetic_classifier['Precision'], marker='o', linestyle='--', label='synthetic Precision' ,color='orange')

# Line Plot for Recall
plt.plot(classes, initial_classifier['Recall'], marker='o', linestyle='-', label='Initial Recall' , color='gray')
plt.plot(classes, synthetic_classifier['Recall'], marker='o', linestyle='--', label='synthetic Recall', color="black")

# Line Plot for F1-Score
plt.plot(classes, initial_classifier['F1-Score'], marker='o', linestyle='-', label='Initial F1-Score', color = 'red')
plt.plot(classes, synthetic_classifier['F1-Score'], marker='o', linestyle='--', label='synthetic F1-Score', color="brown")

plt.title('real data vs synthetic ')
plt.xlabel('Classes')
plt.ylabel('Metric Value')
plt.legend()
plt.grid(True)
plt.savefig('real vs generated', bbox_inches='tight')
plt.show()
```
