```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data for both plots
trials1 = np.array([1, 2, 3, 4, 5])  # Numeric x-axis for plotting
accuracies1 = np.array([62.00, 75.33, 80.33, 83.33, 85.67])
accuracy_stds1 = np.array([7.55, 2.08, 1.53, 0.58, 1.53])

trials2 = np.array([1, 2, 3, 4, 5])  # Same numeric x-axis for second plot
accuracies2 = np.array([62.00, 71.67, 74.67, 75.00, 78.33])
accuracy_stds2 = np.array([7.55, 3.06, 3.21, 1.00, 2.52])

# Set the color palette
colors = sns.color_palette("Set2", 2)

# Seaborn style
sns.set(style="whitegrid")

# Creating subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 7))

# Function to plot with shaded standard deviation
def plot_with_std(ax, x, y, std, color, label):
    ax.plot(x, y, '-o', color=color, label=label)
    ax.fill_between(x, y - std, y + std, color=color, alpha=0.2)


# Plotting accuracy with standard deviation on the first subplot
plot_with_std(ax1, trials1, accuracies1, accuracy_stds1, colors[0], 'Real MRI Images: Accuracy ± STD')
ax1.set_title('Real MRI Images', fontsize=19)
ax1.set_xlabel('\nTotal Number of Training Images', fontsize=19)
ax1.set_ylabel('Accuracy (%)', fontsize=19)
ax1.set_xticks(trials1)
ax1.set_xticklabels(['72', '144', '216', '360', '720'], fontsize=14)
ax1.legend(fontsize=14)
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
ax1.set_ylim(50, 100)

# Plotting accuracy with standard deviation on the second subplot
plot_with_std(ax2, trials2, accuracies2, accuracy_stds2, colors[1], 'Real & Synthetic Mix: Accuracy ± STD')
ax2.set_title('Real & Synthetic Mix', fontsize=19)
ax2.set_xlabel('\nNumber of Training Images (Real & Synthetic)', fontsize=19)
ax2.set_ylabel('\n\nAccuracy (%)', fontsize=19)
ax2.set_xticks(trials2)
ax2.set_xticklabels(['(r=72, s=0)', '(r=72, s=72)', '(r=72, s=144)', '(r=72, s=288)', '(r=72, s=648)'], fontsize=14)
ax2.legend(fontsize=14)  # Legend for second subplot
ax2.grid(True, which='both', linestyle='--', linewidth=0.5)
ax2.set_ylim(50, 100)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data for the combined plot
trials = ['72', '144', '216', '360', '720']  # Common x-axis categories
accuracies1 = [62.00, 75.33, 80.33, 83.33, 85.67]
accuracy_stds1 = [7.55, 2.08, 1.53, 0.58, 1.53]
accuracies2 = [62.00, 71.67, 74.67, 75.00, 78.33]
accuracy_stds2 = [7.55, 3.06, 3.21, 1.00, 2.52]

# Set the color palette
colors = sns.color_palette("Set2", 2)

# Seaborn style
sns.set(style="whitegrid")

# Creating the plot
plt.figure(figsize=(10, 6))

# Adjusting the x-values for each dataset to plot them side by side
x1 = np.arange(len(trials))  # Original x-values for the first dataset
x2 = x1  # Slightly offset x-values for the second dataset

# Plotting accuracy for the first dataset
plt.errorbar(x1, accuracies1, yerr=accuracy_stds1, fmt='-o', capsize=5, capthick=2, ecolor=colors[0], color=colors[0], label='Real MRI Images: Accuracy ± STD')

# Plotting accuracy for the second dataset
plt.errorbar(x2, accuracies2, yerr=accuracy_stds2, fmt='-^', capsize=5, capthick=2, ecolor=colors[1], color=colors[1], label='Real & Synthetic Mix: Accuracy ± STD')

# Titles and Labels
plt.title('Comparison of Accuracy Between Real MRI Images and Real & Synthetic Mix')
plt.xlabel('Number of Training Images')
plt.ylabel('Accuracy (%)')
plt.xticks((x1 + x2) / 2, trials)  # Setting the x-ticks in the middle of each group
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.ylim(50, 100)  # Setting y-axis limit for Accuracy

# Legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
