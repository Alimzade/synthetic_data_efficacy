```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load confusion matrix images
initial_matrices = [
    "confusion_matrix_33.png",
    "confusion_matrix_66.png",
    "confusion_matrix_99.png",
]

fine_tuning_matrices = [
    "confusion_matrix_FineTuning33.png",
    "confusion_matrix_FineTuning66.png",
    "confusion_matrix_FineTuning99.png",
]

def load_image(file_path):
    return np.array(Image.open(file_path))

# Load initial confusion matrices
initial_matrices_data = [load_image(matrix) for matrix in initial_matrices]

# Load fine-tuning confusion matrices
fine_tuning_matrices_data = [load_image(matrix) for matrix in fine_tuning_matrices]

# Create subplots with adjusted layout
fig, axs = plt.subplots(2, 3, figsize=(15, 7), tight_layout=True)

# Plot the initial confusion matrices side by side
for i, (ax, matrix_data) in enumerate(zip(axs[0], initial_matrices_data), 1):
    ax.imshow(matrix_data)
    ax.set_title(f'Initial {i}')
    ax.axis('off')

# Plot the fine-tuning confusion matrices side by side
for i, (ax, matrix_data) in enumerate(zip(axs[1], fine_tuning_matrices_data), 1):
    ax.imshow(matrix_data)
    ax.set_title(f'Fine-tuning {i}')
    ax.axis('off')

#plt.savefig('confusion_matrix_allseeds.png')
plt.show()
```
